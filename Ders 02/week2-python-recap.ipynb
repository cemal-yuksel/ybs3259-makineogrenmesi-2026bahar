````python
# ğŸ—‚ï¸ GRUPLAMA SENARYO 1: Genre bazÄ±nda rating analizi
# "Hangi tÃ¼r iÃ§erikler daha kaliteli?"

print("ğŸ—‚ï¸ SENARYO 1: TÃ¼r (Genre) BazÄ±nda Rating Analizi")
print("="*80)

# Genre'ye gÃ¶re grupla ve ortalama rating hesapla
genre_ratings = df.groupby('genre')['rating'].agg([
    ('Ortalama_Rating', 'mean'),
    ('Ä°Ã§erik_SayÄ±sÄ±', 'count'),
    ('Max_Rating', 'max'),
    ('Min_Rating', 'min')
]).round(2)

# Rating'e gÃ¶re sÄ±rala (en yÃ¼ksekten en dÃ¼ÅŸÃ¼ÄŸe)
genre_ratings = genre_ratings.sort_values('Ortalama_Rating', ascending=False)

print("\nğŸ“Š TÃœR BAZINDA RATING ANALÄ°ZÄ°:")
print(genre_ratings)

# Ä°lk 5 tÃ¼rÃ¼ vurgula
print("\nğŸŒŸ EN YÃœKSEK RATING'LI 5 TÃœR:")
top_5_genres = genre_ratings.head(5)
for idx, row in top_5_genres.iterrows():
    print(f"   {idx:15} â†’ Ort: {row['Ortalama_Rating']:.2f} â­ ({int(row['Ä°Ã§erik_SayÄ±sÄ±'])} iÃ§erik)")
````

````python
# ğŸ—‚ï¸ GRUPLAMA SENARYO 2: Type (Film vs Dizi) karÅŸÄ±laÅŸtÄ±rmasÄ±
# "Filmler mi diziler mi daha kaliteli?"

print("ğŸ—‚ï¸ SENARYO 2: Film vs Dizi KarÅŸÄ±laÅŸtÄ±rmasÄ±")
print("="*80)

# Type'a gÃ¶re grupla ve detaylÄ± analiz
type_analysis = df.groupby('type').agg({
    'rating': ['mean', 'median', 'std', 'min', 'max'],
    'release_year': ['min', 'max', 'mean'],
    'title': 'count'
}).round(2)

print("\nğŸ“Š FÄ°LM vs DÄ°ZÄ° ANALÄ°ZÄ°:")
print(type_analysis)

# Daha anlaÅŸÄ±lÄ±r format
print("\n" + "="*80)
for content_type in df['type'].unique():
    subset = df[df['type'] == content_type]
    print(f"\nğŸ¬ {content_type.upper()} Ä°STATÄ°STÄ°KLERÄ°:")
    print(f"   Toplam iÃ§erik: {len(subset)}")
    print(f"   Ortalama rating: {subset['rating'].mean():.2f}")
    print(f"   Rating aralÄ±ÄŸÄ±: {subset['rating'].min():.2f} - {subset['rating'].max():.2f}")
    print(f"   YÄ±l aralÄ±ÄŸÄ±: {int(subset['release_year'].min())} - {int(subset['release_year'].max())}")
````

````python
# ğŸ—‚ï¸ GRUPLAMA SENARYO 3: Ã‡oklu gruplama (Genre + Type)
# "Her tÃ¼rÃ¼n film ve dizi versiyonlarÄ± nasÄ±l performans gÃ¶steriyor?"

print("ğŸ—‚ï¸ SENARYO 3: Ä°ki Seviye Gruplama (Genre + Type)")
print("="*80)

# Genre ve Type'a gÃ¶re Ã§ift seviyeli gruplama
multi_group = df.groupby(['genre', 'type'])['rating'].agg([
    ('Ortalama', 'mean'),
    ('Adet', 'count')
]).round(2)

# Sadece en az 10 iÃ§eriÄŸi olan kombinasyonlarÄ± gÃ¶ster
multi_group_filtered = multi_group[multi_group['Adet'] >= 10]
multi_group_filtered = multi_group_filtered.sort_values('Ortalama', ascending=False)

print("\nğŸ“Š TÃœR + Ä°Ã‡ERÄ°K TÄ°PÄ° KOMBÄ°NASYONLARI (En az 10 iÃ§erik):")
print(multi_group_filtered.head(15))

# En popÃ¼ler kombinasyonlar
print("\nğŸ”¥ EN POPÃœLER 5 KOMBÄ°NASYON:")
top_combos = multi_group.sort_values('Adet', ascending=False).head(5)
print(top_combos)
````

````python
# ğŸ—‚ï¸ GRUPLAMA SENARYO 4: YÄ±l bazÄ±nda trend analizi
# "Ä°Ã§erik Ã¼retimi ve kalitesi yÄ±llar iÃ§inde nasÄ±l deÄŸiÅŸti?"

print("ğŸ—‚ï¸ SENARYO 4: YÄ±l BazÄ±nda Trend Analizi")
print("="*80)

# YÄ±l bazÄ±nda iÃ§erik sayÄ±sÄ± ve ortalama rating
yearly_trends = df.groupby('release_year').agg({
    'title': 'count',
    'rating': 'mean'
}).round(2)

yearly_trends.columns = ['Ä°Ã§erik_SayÄ±sÄ±', 'Ortalama_Rating']

# Son 10 yÄ±lÄ± gÃ¶ster
print("\nğŸ“ˆ SON 10 YIL TRENDÄ°:")
recent_years = yearly_trends.tail(10)
print(recent_years)

# En Ã§ok iÃ§erik Ã¼retilen 5 yÄ±l
print("\nğŸ”¥ EN Ã‡OK Ä°Ã‡ERÄ°K ÃœRETÄ°LEN 5 YIL:")
top_years = yearly_trends.sort_values('Ä°Ã§erik_SayÄ±sÄ±', ascending=False).head(5)
for year, row in top_years.iterrows():
    print(f"   {int(year)} â†’ {int(row['Ä°Ã§erik_SayÄ±sÄ±'])} iÃ§erik (Ort rating: {row['Ortalama_Rating']:.2f})")
````

````python
# ğŸ”„ DÃ–NÃœÅTÃœRME SENARYO 1: Rating kategorisi oluÅŸturma
# Rating'i Ã¼Ã§ kategoriye ayÄ±ralÄ±m: YÃ¼ksek, Orta, DÃ¼ÅŸÃ¼k

print("ğŸ”„ SENARYO 1: Rating Kategorisi OluÅŸturma")
print("="*80)

# Fonksiyon tanÄ±mlama
def rating_kategorisi(rating):
    if rating >= 7.5:
        return 'YÃ¼ksek â­â­â­'
    elif rating >= 5.5:
        return 'Orta â­â­'
    else:
        return 'DÃ¼ÅŸÃ¼k â­'

# Yeni sÃ¼tun oluÅŸtur (apply fonksiyonu ile)
df['rating_kategori'] = df['rating'].apply(rating_kategorisi)

print("âœ… Yeni sÃ¼tun 'rating_kategori' oluÅŸturuldu!\n")

# DaÄŸÄ±lÄ±mÄ± gÃ¶relim
print("ğŸ“Š RATING KATEGORÄ° DAÄILIMI:")
kategori_dagilim = df['rating_kategori'].value_counts()
for kategori, adet in kategori_dagilim.items():
    yuzde = (adet / len(df)) * 100
    print(f"   {kategori:20} : {adet:4d} iÃ§erik ({yuzde:5.2f}%)")

# Ã–rnekler
print("\nğŸ’¡ Ã–rnek Veriler:")
print(df[['title', 'rating', 'rating_kategori']].head(10))
````

````python
# ğŸ”„ DÃ–NÃœÅTÃœRME SENARYO 2: Ä°Ã§erik yaÅŸÄ± hesaplama
# "Bu iÃ§erik kaÃ§ yaÅŸÄ±nda?"

print("ğŸ”„ SENARYO 2: Ä°Ã§erik YaÅŸÄ± Hesaplama")
print("="*80)

# 2026'dan release_year'Ä± Ã§Ä±kararak yaÅŸ hesapla
current_year = 2026
df['content_age'] = current_year - df['release_year']

print("âœ… Yeni sÃ¼tun 'content_age' oluÅŸturuldu!\n")

# Ä°statistikler
print("ğŸ“Š Ä°Ã‡ERÄ°K YAÅ Ä°STATÄ°STÄ°KLERÄ°:")
print(f"   Ortalama yaÅŸ: {df['content_age'].mean():.1f} yÄ±l")
print(f"   En yeni iÃ§erik: {df['content_age'].min()} yÄ±l")
print(f"   En eski iÃ§erik: {df['content_age'].max()} yÄ±l")

# YaÅŸ kategorileri
print("\nğŸ—“ï¸ YAÅ KATEGORÄ°LERÄ°:")
print(f"   0-5 yÄ±l (Yeni): {len(df[df['content_age'] <= 5])} iÃ§erik")
print(f"   6-10 yÄ±l (Orta): {len(df[(df['content_age'] > 5) & (df['content_age'] <= 10)])} iÃ§erik")
print(f"   10+ yÄ±l (Eski): {len(df[df['content_age'] > 10])} iÃ§erik")

# Yeni iÃ§eriklerin rating ortalamasÄ±
yeni_icerik_rating = df[df['content_age'] <= 5]['rating'].mean()
eski_icerik_rating = df[df['content_age'] > 10]['rating'].mean()

print(f"\nğŸ’¡ YORUM:")
print(f"   Yeni iÃ§eriklerin ort rating: {yeni_icerik_rating:.2f}")
print(f"   Eski iÃ§eriklerin ort rating: {eski_icerik_rating:.2f}")
````

````python
# ğŸ”„ DÃ–NÃœÅTÃœRME SENARYO 3: String iÅŸlemleri
# BaÅŸlÄ±k manipÃ¼lasyonu ve temizleme

print("ğŸ”„ SENARYO 3: String Ä°ÅŸlemleri")
print("="*80)

# BaÅŸlÄ±klarÄ± bÃ¼yÃ¼k harfe Ã§evir
df['title_upper'] = df['title'].str.upper()

# BaÅŸlÄ±k uzunluÄŸunu hesapla
df['title_length'] = df['title'].str.len()

# BaÅŸlÄ±kta sayÄ± var mÄ± kontrol et
df['has_number'] = df['title'].str.contains(r'\d', regex=True)

print("âœ… Yeni string sÃ¼tunlarÄ± oluÅŸturuldu!\n")

# Ä°statistikler
print("ğŸ“Š STRING ANALÄ°ZÄ°:")
print(f"   Ortalama baÅŸlÄ±k uzunluÄŸu: {df['title_length'].mean():.1f} karakter")
print(f"   En uzun baÅŸlÄ±k: {df['title_length'].max()} karakter")
print(f"   En kÄ±sa baÅŸlÄ±k: {df['title_length'].min()} karakter")
print(f"   BaÅŸlÄ±ÄŸÄ±nda sayÄ± olan iÃ§erik: {df['has_number'].sum()} adet")

# Ã–rnekler
print("\nğŸ’¡ Ã–RNEKLER:")
sample = df[['title', 'title_upper', 'title_length', 'has_number']].head(10)
print(sample)

# En uzun baÅŸlÄ±klÄ± iÃ§erikler
print("\nğŸ“ EN UZUN BAÅLIKLI 5 Ä°Ã‡ERÄ°K:")
longest = df.nlargest(5, 'title_length')[['title', 'title_length', 'type']]
for idx, row in longest.iterrows():
    print(f"   ({row['title_length']} kar.) {row['title']}")
````

## ğŸ“Š **DÃ–RDÃœNCÃœ GRAFÄ°ÄÄ°MÄ°Z: Film vs Dizi OranÄ± (Pie Chart)**

**ğŸ’¼ YBS Sorusu:** Ä°Ã§erik portfÃ¶yÃ¼mÃ¼zÃ¼n daÄŸÄ±lÄ±mÄ± nasÄ±l? Film mi dizi mi aÄŸÄ±rlÄ±klÄ±?

**Pasta grafiÄŸi** oranlarÄ± ve yÃ¼zdeleri gÃ¶stermek iÃ§in mÃ¼kemmeldir.

````python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# =========================
# 1) Veri hazÄ±rlÄ±ÄŸÄ± (saÄŸlamlaÅŸtÄ±rma)
# =========================
df_plot = df.copy()

# type alanÄ±nÄ± temizle: NaN -> 'Unknown', strip, standardize
df_plot['type'] = (
    df_plot['type']
    .astype(str)
    .str.strip()
    .replace({'nan': 'Unknown', 'None': 'Unknown', '': 'Unknown'})
)

# Ä°steÄŸe baÄŸlÄ±: Movie/TV Show dÄ±ÅŸÄ±ndakileri "DiÄŸer"e topla (Netflix veri setlerinde faydalÄ±)
main_map = {'Movie': 'Movie', 'TV Show': 'TV Show'}
df_plot['type_clean'] = df_plot['type'].where(df_plot['type'].isin(main_map), other='DiÄŸer')

type_counts = df_plot['type_clean'].value_counts()

# EÄŸer tamamen tek kategori varsa (edge case), pie yerine bar daha mantÄ±klÄ± olur.
if len(type_counts) < 2:
    print("UyarÄ±: Pasta grafik iÃ§in en az 2 kategori gerekir. Mevcut daÄŸÄ±lÄ±m:")
    print(type_counts)
else:
    # =========================
    # 2) Profesyonel gÃ¶rselleÅŸtirme (Donut + dinamik explode + dinamik renk)
    # =========================
    fig, ax = plt.subplots(figsize=(10, 7))

    # Matplotlib'in varsayÄ±lan renk dÃ¶ngÃ¼sÃ¼nden dinamik renk al (kategoriler kaÃ§sa o kadar)
    base_colors = plt.rcParams['axes.prop_cycle'].by_key()['color']
    colors = (base_colors * ((len(type_counts) // len(base_colors)) + 1))[:len(type_counts)]

    # Hafif ayrÄ±ÅŸtÄ±rma: her dilim iÃ§in aynÄ± explode (dinamik uzunluk!)
    explode = [0.05] * len(type_counts)

    total = type_counts.sum()

    def autopct_fmt(pct):
        count = int(round(pct * total / 100.0))
        return f"{pct:.1f}%\n({count})"

    wedges, texts, autotexts = ax.pie(
        type_counts.values,
        labels=type_counts.index,
        autopct=autopct_fmt,
        colors=colors,
        explode=explode,
        startangle=90,
        counterclock=False,
        shadow=True,
        textprops={'fontsize': 12, 'fontweight': 'bold'}
    )

    # Donut efekti (merkeze beyaz daire)
    centre_circle = plt.Circle((0, 0), 0.55, fc='white')
    ax.add_artist(centre_circle)

    # YÃ¼zde/adet metinlerini iyileÅŸtir
    for t in autotexts:
        t.set_color('black')
        t.set_fontsize(11)
        t.set_fontweight('bold')

    ax.set_title(
        'Netflix Ä°Ã§erik PortfÃ¶yÃ¼: Film vs Dizi (ve DiÄŸer)',
        fontsize=16,
        fontweight='bold',
        pad=18
    )

    # Legend: saÄŸda, temiz
    ax.legend(
        wedges,
        [f"{k} â€” {v} iÃ§erik" for k, v in type_counts.items()],
        title='Ä°Ã§erik Tipi',
        loc='center left',
        bbox_to_anchor=(1.02, 0.5),
        fontsize=11
    )

    ax.set_aspect('equal')  # daireyi bozmamak iÃ§in
    plt.tight_layout()
    plt.show()

# =========================
# 3) DetaylÄ± analiz (daha dÃ¼zenli)
# =========================
print("\nğŸ“Š Ä°Ã‡ERÄ°K TÄ°PÄ° DAÄILIMI:")
for content_type, count in type_counts.items():
    percentage = (count / len(df_plot)) * 100
    print(f"   {content_type:10} : {count:5d} iÃ§erik  (%{percentage:5.1f})")

# Rating karÅŸÄ±laÅŸtÄ±rmasÄ± (NaN gÃ¼venli)
print("\nâ­ RATING KARÅILAÅTIRMASI (ortalama):")
if 'rating' in df_plot.columns:
    # rating sayÄ±sal deÄŸilse dÃ¶nÃ¼ÅŸtÃ¼rmeyi deneyelim (hata vermesin)
    rating_num = pd.to_numeric(df_plot['rating'], errors='coerce')
    df_plot['rating_num'] = rating_num

    for content_type in df_plot['type_clean'].unique():
        avg_rating = df_plot.loc[df_plot['type_clean'] == content_type, 'rating_num'].mean()
        if np.isnan(avg_rating):
            print(f"   {content_type:10} : Ortalama hesaplanamadÄ± (sayÄ±sal deÄŸil/eksik)")
        else:
            print(f"   {content_type:10} : Ortalama {avg_rating:.2f}")
else:
    print("   'rating' sÃ¼tunu bulunamadÄ±.")

# Yorumu daha saÄŸlam kur (value_counts sÄ±ralamasÄ± deÄŸiÅŸebilir)
dominant = type_counts.idxmax()
print(f"\nğŸ’¡ YORUM: PortfÃ¶yde baskÄ±n iÃ§erik tipi: {dominant} (n={type_counts.max()}).")
````

## ğŸ”¥ **DÃ–RDÃœNCÃœ GRAFÄ°K: Heatmap - Korelasyon Matrisi**

**Heatmap** (IsÄ± HaritasÄ±), sayÄ±sal deÄŸiÅŸkenler arasÄ±ndaki **iliÅŸkiyi** gÃ¶sterir. 

**Korelasyon:**
- **+1**: MÃ¼kemmel pozitif iliÅŸki (biri artarsa diÄŸeri de artar)
- **0**: Ä°liÅŸki yok
- **-1**: MÃ¼kemmel negatif iliÅŸki (biri artarsa diÄŸeri azalÄ±r)

    B --> D[ğŸ’¼ Ä°ÅŸ KararÄ±:<br/>Bu deÄŸiÅŸkenler<br/>birlikte hareket eder]

````python
# ğŸ”¥ SEABORN GRAFÄ°K 4: Korelasyon Heatmap

# Sadece sayÄ±sal sÃ¼tunlarÄ± seÃ§ ve korelasyon hesapla
numeric_df = df[['rating', 'release_year', 'content_age']].copy()
correlation_matrix = numeric_df.corr()

# Figure oluÅŸtur
plt.figure(figsize=(10, 8))

# Heatmap Ã§iz
# annot=True: SayÄ±larÄ± gÃ¶ster
# fmt='.2f': VirgÃ¼lden sonra 2 basamak
# cmap: Renk paleti
# square=True: HÃ¼creler kare olsun
# linewidths: HÃ¼creler arasÄ± Ã§izgi
sns.heatmap(correlation_matrix, annot=True, fmt='.2f', 
            cmap='RdYlGn', center=0, square=True, 
            linewidths=2, cbar_kws={"shrink": 0.8},
            vmin=-1, vmax=1)

plt.title('Netflix: DeÄŸiÅŸkenler ArasÄ± Korelasyon HaritasÄ±', 
          fontsize=16, fontweight='bold', pad=20)

plt.tight_layout()
plt.show()

# KorelasyonlarÄ± yorumla
print(f"\nğŸ”¥ KORELASYON ANALÄ°ZÄ°:")
print(f"="*70)
print(f"\nğŸ“Š Korelasyon Matrisi:")
print(correlation_matrix)

print(f"\nğŸ’¡ YORUM:")
print(f"   ğŸŸ¢ YEÅIL (Pozitif) = Biri artar, diÄŸeri de artar")
print(f"   ğŸ”´ KIRMIZI (Negatif) = Biri artar, diÄŸeri azalÄ±r")
print(f"   ğŸŸ¡ SARI (SÄ±fÄ±ra yakÄ±n) = Ä°liÅŸki yok")

# En gÃ¼Ã§lÃ¼ korelasyonu bul (diyagonal hariÃ§)
mask = correlation_matrix != 1.0  # Kendisiyle korelasyon 1.0'Ä± Ã§Ä±kar
max_corr = correlation_matrix[mask].abs().max()
max_loc = correlation_matrix[mask].abs().idxmax().idxmax()

print(f"\nğŸ¯ EN GÃœÃ‡LÃœ Ä°LÄ°ÅKÄ°:")
val1, val2 = correlation_matrix[mask].abs().stack().idxmax()
corr_val = correlation_matrix.loc[val1, val2]
print(f"   {val1} â†”ï¸ {val2}")
print(f"   Korelasyon: {corr_val:.3f}")

if abs(corr_val) > 0.7:
    print(f"   â†’ GÃ¼Ã§lÃ¼ iliÅŸki! ğŸ’ª")
elif abs(corr_val) > 0.4:
    print(f"   â†’ Orta seviye iliÅŸki ğŸ‘")
else:
    print(f"   â†’ ZayÄ±f iliÅŸki ğŸ¤·")
````

````python
# ğŸ“ˆ PLOTLY GRAFÄ°K 1: YÄ±llÄ±k Ä°Ã§erik Trendi (Ä°nteraktif Line Chart)

# YÄ±llara gÃ¶re iÃ§erik sayÄ±sÄ±
yearly_data = df.groupby('release_year').size().reset_index(name='count')

# Plotly Express ile TEK SATIRDA interaktif grafik!
fig = px.line(yearly_data, x='release_year', y='count',
              title='Netflix: YÄ±llara GÃ¶re Ä°Ã§erik Ãœretim Trendi',
              labels={'release_year': 'YÄ±l', 'count': 'Ä°Ã§erik SayÄ±sÄ±'},
              markers=True)  # Nokta iÅŸaretleri ekle

# Ã‡izgi renkini ve kalÄ±nlÄ±ÄŸÄ±nÄ± ayarla
fig.update_traces(line_color='#F4A261', line_width=3, marker_size=8)

# Layout ayarlarÄ±
fig.update_layout(
    hovermode='x unified',  # Hover bilgisi tÃ¼m grafikte gÃ¶rÃ¼nsÃ¼n
    template='plotly_white',  # Beyaz arka plan
    height=500,
    font=dict(size=12),
    title_font=dict(size=18, color='#2C3E50')
)

# GrafiÄŸi gÃ¶ster
fig.show()

# Ä°statistikler
peak_year = yearly_data.loc[yearly_data['count'].idxmax(), 'release_year']
peak_count = yearly_data['count'].max()

print(f"\nğŸ“ˆ YILLIK TREND ANALÄ°ZÄ°:")
print(f"   En fazla iÃ§erik Ã¼retilen yÄ±l: {peak_year} ({peak_count} iÃ§erik)")
print(f"   Ä°lk yÄ±l: {yearly_data['release_year'].min()}")
print(f"   Son yÄ±l: {yearly_data['release_year'].max()}")
print(f"   Toplam yÄ±l aralÄ±ÄŸÄ±: {yearly_data['release_year'].max() - yearly_data['release_year'].min()} yÄ±l")

print(f"\nğŸ’¡ Ä°NTERAKTÄ°F Ã–ZELLÄ°KLER:")
print(f"   ğŸ” Scroll ile ZOOM yap")
print(f"   â„¹ï¸ NoktalarÄ±n Ã¼zerine gel, detaylarÄ± gÃ¶r")
print(f"   ğŸ“¸ SaÄŸ Ã¼stteki ğŸ“· ikonuna tÄ±kla, PNG indir")
print(f"   ğŸ”„ Ã‡ift tÄ±kla, zoom'u sÄ±fÄ±rla")
````

## ğŸ¨ **Ä°KÄ°NCÄ° GRAFÄ°K: Genre-Rating Ä°liÅŸkisi (Scatter Plot)**

**ğŸ’¼ YBS Sorusu:** Hangi tÃ¼rlerin rating'i yÃ¼ksek ve popÃ¼ler?

**Scatter Plot (DaÄŸÄ±lÄ±m GrafiÄŸi):**
- X ekseni: Ortalama rating
- Y ekseni: Ä°Ã§erik sayÄ±sÄ±
- Nokta boyutu: Toplam rating puanÄ±
- Renk: TÃ¼r kategorisi

**âœ¨ Ä°nteraktif:** Hover ile tÃ¼r adÄ±nÄ± ve detaylarÄ± gÃ¶r!

````python
# ğŸ¨ PLOTLY GRAFÄ°K 2: Genre-Rating Ä°liÅŸkisi (Bubble Scatter)

# Top 15 genre seÃ§
top_15_genres = df['genre'].value_counts().head(15).index
df_top15 = df[df['genre'].isin(top_15_genres)]

# Genre bazÄ±nda istatistikler
genre_stats = df_top15.groupby('genre').agg({
    'rating': ['mean', 'count', 'sum']
}).reset_index()
genre_stats.columns = ['genre', 'avg_rating', 'content_count', 'total_rating']

# Bubble Scatter Plot
fig = px.scatter(genre_stats, 
                 x='avg_rating', 
                 y='content_count',
                 size='total_rating',  # Balon boyutu
                 color='genre',  # Renk kategorisi
                 hover_name='genre',  # Hover'da gÃ¶ster
                 hover_data={'avg_rating': ':.2f', 'content_count': True, 'total_rating': ':.0f'},
                 title='Netflix: Genre-Rating Ä°liÅŸkisi (Bubble Chart)',
                 labels={'avg_rating': 'Ortalama Rating', 
                         'content_count': 'Ä°Ã§erik SayÄ±sÄ±',
                         'total_rating': 'Toplam Rating PuanÄ±'},
                 size_max=60)  # Maksimum balon boyutu

# Layout
fig.update_layout(
    template='plotly_white',
    height=600,
    showlegend=True,
    legend=dict(title='TÃ¼r (Genre)', font=dict(size=10)),
    font=dict(size=12)
)

fig.show()

# En iyi performans gÃ¶steren tÃ¼rÃ¼ bul
best_genre = genre_stats.loc[genre_stats['avg_rating'].idxmax()]

print(f"\nğŸ¯ GENRE-RATING ANALÄ°ZÄ°:")
print(f"   En yÃ¼ksek ortalama rating: {best_genre['genre']} ({best_genre['avg_rating']:.2f})")
print(f"   En fazla iÃ§erik: {genre_stats.loc[genre_stats['content_count'].idxmax(), 'genre']} "
      f"({genre_stats['content_count'].max()} iÃ§erik)")

print(f"\nğŸ’¡ BUBBLE CHART YORUMU:")
print(f"   ğŸ“ SAÄ ÃœST kÃ¶ÅŸe = Hem popÃ¼ler, hem kaliteli (Ä°DEAL!)")
print(f"   ğŸ“ SOL ÃœST = PopÃ¼ler ama dÃ¼ÅŸÃ¼k rating")
print(f"   ğŸ“ SAÄ ALT = YÃ¼ksek rating ama az iÃ§erik")
print(f"   ğŸ”µ BÃœYÃœK balonlar = Toplam rating yÃ¼ksek")
````

## ğŸŒ **ÃœÃ‡ÃœNCÃœ GRAFÄ°K: Ãœlkelere GÃ¶re Ä°Ã§erik (Ä°nteraktif Bar Chart)**
